## DataFrames:

### BaseServiçosPrestados.xlsx

```python
import pandas as pd

df_baseServicos = pd.read_excel('BaseServiçosPrestados.xlsx')
df_baseServicos
```

### CadastroClientes.csv

```python
df_cadastroClientes = pd.read_csv('CadastroClientes.csv',sep=';')

df_cadastroClientes
```

### CadastroFuncionarios.csv

```python
df_CadastroFuncionario = pd.read_csv('CadastroFuncionarios.csv',sep=';',decimal=',')

df_CadastroFuncionario
```

## Análises:

### 1 - Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? 

```python
df_CadastroFuncionario['Salario Total'] = (df_CadastroFuncionario['Salario Base'] + df_CadastroFuncionario['Impostos'] + df_CadastroFuncionario['Beneficios'] + df_CadastroFuncionario['VT'] + df_CadastroFuncionario['VR'])
FolhaSalarial = df_CadastroFuncionario['Salario Total'].sum()
display(df_CadastroFuncionario)
print(f'O Valor Total da Folha Salarial é de R${FolhaSalarial:,}')
```

### 2 - Qual foi o faturamento da empresa?

```python
df_faturamento = df_baseServicos[['ID Cliente','Tempo Total de Contrato (Meses)']].merge(df_cadastroClientes[['ID Cliente', 'Valor Contrato Mensal']],on='ID Cliente')
df_faturamento['Faturamento Total'] = df_faturamento['Valor Contrato Mensal'] * df_faturamento['Tempo Total de Contrato (Meses)']
faturamentoEmpresa = df_faturamento['Faturamento Total'].sum()
display(df_faturamento)

print(f'O faturamento total da empresa é de: R${faturamentoEmpresa:,}')
```

### 3 - Qual o % de funcionários que já fechou algum contrato?

```python
qtde_funcionarios_fecharam_servicos = len(df_baseServicos['ID Funcionário'].unique())
qtde_funcionarios_totais = len(df_CadastroFuncionario['ID Funcionário'])

print(f'O percentual de funcionários que fecharam serviço é de: {(qtde_funcionarios_fecharam_servicos/qtde_funcionarios_totais):.2%}%')
```

### 4 - Calcule o total de contratos que cada área da empresa já fechou

```python
df_contratosArea = df_baseServicos[['ID Funcionário','Codigo do Servico']].merge(df_CadastroFuncionario[['ID Funcionário','Area']],on='ID Funcionário')

display(df_contratosArea)
area_counts = df_contratosArea['Area'].value_counts()

for area, contratos in area_counts.items():
    print(f'O setor {area} possui um total de {contratos} contratos.')
```

### 5 - Calcule o total de funcionários por área

```python
contagem = df_CadastroFuncionario['Area'].value_counts()

for area, quant in contagem.items():
    print(f'A área da {area} tem {quant} funcionarios.')
```

### 6 - Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

```python
ticketMedio = df_cadastroClientes['Valor Contrato Mensal'].mean()
print(f'O Valor do Ticket Médio da Empresa é de: R${ticketMedio:,.2f}')
```
